```python
import matplotlib.pyplot as plt
import numpy as np
import glob, os
```

```python
x = range(25,30+1)
times = []
for i in x:
  with open(f"unif/times-unif-{i}.txt", 'r') as f:
    line = f.read().splitlines()[-2]
    # print(line)
    times.append(float(line.split(' ')[6]))
plt.plot(list(x), times)
plt.yscale('log')
plt.xlabel('exponent input size (2^x)')
plt.ylabel('time (s)')
plt.title('Time to construct SA from uniform input.')
plt.savefig('unif/unif.jpg', dpi=150)
  
```
